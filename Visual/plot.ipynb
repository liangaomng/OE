```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.load("../Model_out/2_29_expr_1/seed_10_modes32_alpha_0.1_losses_with_epoch.npz",allow_pickle=True)
```

```python
for key in data.files:
    print(key)
 
```

```python
print(data["local_metric"])
```

```python
global_metric_dict = data['global_metric'].item()
local_metric_dict = data['local_metric'].item()
print(local_metric_dict.keys())
```

```python

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].scatter(global_metric_dict["psnr"],local_metric_dict["psnr"],alpha=0.2,label="alpha=0.1")
ax[0].set_xlabel("Global psnr with epochs")
ax[0].set_ylabel("Local psnr with epochs")
ax[1].scatter(global_metric_dict["ssim"],local_metric_dict["ssim"],alpha=0.2,label="alpha=0.1")
ax[0].legend()
ax[1].legend()
ax[1].set_xlabel("Global ssim with epochs")
ax[1].set_ylabel("Local ssim with epochs")

```
